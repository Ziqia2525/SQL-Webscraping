```python
import pandas as pd
import numpy as np
import sqlite3
import requests
```

Extract

```python

# Extract data from SWAPI

def get_swapi_data(endpoint):
    url = f"https://swapi.dev/api/{endpoint}/"
    response = requests.get(url)
    data = response.json()
    return data

def get_all_swapi_data(endpoint):
    all_data = []
    url = f"https://swapi.dev/api/{endpoint}/"
    
    while url:
        response = requests.get(url)
        data = response.json()
        all_data.extend(data['results'])
        url = data['next']
    
    return all_data


people_data = get_all_swapi_data('people')
films_data = get_all_swapi_data('films')
vehicles_data = get_all_swapi_data('vehicles')
planets_data = get_all_swapi_data('planets')
starships_data = get_all_swapi_data('starships')


df_people = pd.DataFrame(people_data)
df_films = pd.DataFrame(films_data)
df_vehicles = pd.DataFrame(vehicles_data)
df_planets = pd.DataFrame(planets_data)
df_starships = pd.DataFrame(starships_data)

# This code sometimes runs smoothly. Sometimes it will report an error, but it will succeed after trying to rerun it once or twice.

```

Transform

```python
df_people_cleaned = df_people[[
    'name',
    'height',
    'mass',
    'hair_color',
    'skin_color',
    'eye_color',
    'birth_year',
    'gender',
    'homeworld',
    'films',
    'species',
    'starships',
    'url'
]]

df_people_cleaned.loc[:, 'species'] = df_people_cleaned['species'].astype(str)
df_people_cleaned.loc[:, 'films'] = df_people_cleaned['films'].astype(str)
df_people_cleaned.loc[:, 'starships'] = df_people_cleaned['starships'].astype(str)
df_people_cleaned = df_people_cleaned.fillna('')
df_people_cleaned.tail()
```

```python
df_planets_cleaned = df_planets[[
    'name',
    'rotation_period',
    'orbital_period',
    'diameter',
    'climate',
    'gravity',
    'terrain',
    'surface_water',
    'population',
    'films',
    'url'
]]

df_planets_cleaned.loc[:, 'films'] = df_planets_cleaned['films'].astype(str)
df_planets_cleaned = df_planets_cleaned.fillna('')
df_planets_cleaned.tail()
```

```python
df_films_cleaned = df_films[[
    'title',
    'episode_id',
    'opening_crawl',
    'director',
    'producer',
    'release_date',
    'url',
]]

df_films_cleaned = df_films_cleaned.fillna('')
df_films_cleaned.tail()
```

```python
df_starships_cleaned = df_starships[[
    'name',
    'model',
    'manufacturer',
    'cost_in_credits',
    'length',
    'max_atmosphering_speed',
    'passengers',
    'starship_class',
    'url',
]]

df_starships_cleaned = df_starships_cleaned.fillna('')
df_starships_cleaned.tail()
```

Load

```python
conn = sqlite3.connect('starwar.db')
c = conn.cursor()


c.execute(f"CREATE TABLE IF NOT EXISTS people ({', '.join(df_people_cleaned.columns)})")
conn.commit()
df_people_cleaned.to_sql('people', conn, if_exists='replace', index = False)

c.execute(f"CREATE TABLE IF NOT EXISTS planets ({', '.join(df_planets_cleaned.columns)})")
conn.commit()
df_planets_cleaned.to_sql('planets', conn, if_exists='replace', index = False)

c.execute(f"CREATE TABLE IF NOT EXISTS films ({', '.join(df_films_cleaned.columns)})")
conn.commit()
df_films_cleaned.to_sql('films', conn, if_exists='replace', index = False)

c.execute(f"CREATE TABLE IF NOT EXISTS starships ({', '.join(df_starships_cleaned.columns)})")
conn.commit()
df_starships_cleaned.to_sql('starships', conn, if_exists='replace', index = False)

```

Analysis

```python
con = sqlite3.connect('starwar.db')
```

```python
# The number of species under different climate type
# INNER JOIN 
# GROUP BY

pd.read_sql(
    '''
        SELECT 
            p2.climate,
            COUNT(DISTINCT p1.species) AS species_count
        FROM 
            people p1
        INNER JOIN 
            planets p2 ON p1.homeworld = p2.url
        WHERE
            p1.species != 'unknown' AND p1.species != ' '
        GROUP BY
            p2.climate
        ORDER BY 
            species_count DESC 
        LIMIT 5
    ''', 
    con
)
```

```python
# The number of characters and average height in each planet
# LEFT JOIN 
# GROUP BY

pd.read_sql(
    '''
        SELECT 
            p2.name,
            count(p1.name) AS characters_count,
            AVG(p1.height) AS average_height
        FROM 
            planets p2
        LEFT JOIN 
            people p1 ON p2.url = p1.homeworld
        WHERE
            p2.name != 'unknown'
        GROUP BY
            p2.name
        ORDER BY 
            characters_count DESC
        LIMIT 5
    ''', 
    con
)
```

```python
# The number of female character in each planet
# INNER JOIN 
# GROUP BY


query = """
SELECT 
    p2.name,
    COUNT(p1.name) AS female_count
FROM 
    people p1
INNER JOIN 
    planets p2 ON p1.homeworld = p2.url
WHERE
    gender = 'female' AND p2.name != 'unknown'
GROUP BY
    p2.name
ORDER BY 
    female_count DESC 
LIMIT 5
"""

pd.read_sql_query(sql=query, con=con)
```

```python
# The number of female character in each film
# LEFT JOIN 
# GROUP BY

pd.read_sql(
    '''
        SELECT 
            f.title,
            f.release_date,
            COUNT(p1.gender) AS female_count
        FROM 
            films f
        LEFT JOIN 
            people p1 ON p1.films LIKE '%' || f.url || '%' 
        WHERE
            p1.gender = 'female'
        GROUP BY
            f.title
        ORDER BY 
            f.release_date ASC 
    ''', 
    con
)
```

```python
# Directors who have directed two or more films
# GROUP BY
# HAVING

query = """
SELECT
    director,
    COUNT(DISTINCT f.title) AS film_count
FROM
    films f
GROUP BY
    director
HAVING
    film_count >= 2
"""

pd.read_sql_query(sql=query, con=con)
```

```python
# The characters who have driven two or more starships
# LEFT JOIN
# GROUP BY
# HAVING

pd.read_sql(
    '''
        SELECT 
            p1.name,
            COUNT(DISTINCT s.name) AS starship_count
        FROM 
            people p1
        LEFT JOIN 
            starships s ON p1.starships LIKE '%' || s.url || '%' 
        GROUP BY
            p1.name
        HAVING
            starship_count >= 2
        ORDER BY
            starship_count DESC;
    ''', 
    con
)
```

```python
# Planets which have black-eye residents
# CTE

pd.read_sql(
    '''
        WITH planet_residents_eye AS (
             SELECT
                 p2.name AS planet_name,
                 p1.name AS resident_name,
                 p1.eye_color AS eye_color
             FROM
                 planets p2
             INNER JOIN
                 people p1 ON p2.url = p1.homeworld 
        )
        SELECT
            planet_name,
            resident_name,
            eye_color
        FROM
            planet_residents_eye
        WHERE
            eye_color = 'black'
    ''', 
    con
)
```

```python
# Number and rank of characters and planets appearing in each film
# CTE

pd.read_sql(
    '''
        WITH film_people AS (
          SELECT
              f.title AS film_title,
              COUNT(p1.name) AS character_count,
              RANK() OVER (ORDER BY COUNT(p1.name) DESC) AS character_rank
          FROM
              films f
          LEFT JOIN
              people p1 ON p1.films LIKE '%' || f.url || '%'
          GROUP BY
              f.title
         ),

        film_planets AS (
          SELECT
              f.title AS film_title,
              COUNT(p2.name) AS planet_count,
              RANK() OVER (ORDER BY COUNT(p2.name) DESC) AS planet_rank
          FROM
              films f
          LEFT JOIN
              planets p2 ON p2.films LIKE '%' || f.url || '%'
          GROUP BY
              f.title
          )
       SELECT
           fp.film_title,
           fp.character_count,
           fp.character_rank,
           fp2.planet_count,
           fp2.planet_rank
      FROM
           film_people fp
      LEFT JOIN
           film_planets fp2 ON fp.film_title = fp2.film_title 
      GROUP BY
           fp.film_title
      ORDER BY
           fp.character_count DESC;
    ''', 
    con
)
```
